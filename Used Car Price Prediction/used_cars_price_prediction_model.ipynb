**LOADING THE DATA SET**


```python
df=pd.read_csv("/Cardekho_Extract.csv")
```

**DROPING COLUMNS WHICH ARE NOT REQUIRED TO BUILD THE MODEL**

```python
df1=df.drop(['Source.Name','web-scraper-order', 'web-scraper-start-url','new-price','seller_type','mileage','engine', 'max_power', 'seats'],axis='columns')
df1.shape
```

```python
df1.head()
```

**FINDING AND REMOVING THE NA VALUES**


```python
df1.isnull().sum()
```

```python
df2=df1.dropna()
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2['full_name'].unique()
```

***In the given data set the name of the car is quite bigger. So by using lambda function taking the first two positions as you can see in the 'car_name' column*****

```python
def get_name(car):
    return car.split(" ")[0]
def get_model(car):
    return car.split(" ")[1]


df2['car_name']=df2['full_name'].apply(lambda x: get_name(x)+' '+get_model(x))
df2.head()
```

**using split function taking only the numerical value from "selling_price" column**

```python
df2['price']=df2['selling_price'].apply(lambda x: x.split(' ')[0])
df2.head()
```

**converting the price to float value**

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['price'].apply(is_float)].head()
```

**Creating a new data frame by avoiding the non-float values**

```python
df3=df2[df2['price'].apply(is_float)]
df3.head()
```

```python
df3.shape
```

```python
df3['price']=df3['price'].map(lambda x: float(x))
```

**Feature engineering- converting the price to lakhs**

```python
df3['price']=df3['price'].apply (lambda x:(x*100000))
df3.head()
```

```python
df3['transmission_type'].unique()
```

```python
df4=df3.drop(['full_name','selling_price','owner_type','transmission_type'],axis='columns')
df4.head()
```

```python
df4['kilometers_driven']=df4['km_driven'].apply(lambda x: x.split(' ')[0])
df4.head()
```

```python
df4.kilometers_driven=[col.replace(',','') for col in df4.kilometers_driven]
df4.head()
```

```python
df4.head()
```

```python
car_count=df4['car_name'].value_counts(ascending=False)
car_count
```

```python
len(car_count[car_count>10])
```

```python
len(car_count[car_count<=10])
```

```python
len(car_count[car_count==1])
```

**DIMENSIONALITY REDUCTION**

```python
car_count_only_one= car_count[car_count==1]
car_count_only_one
```

```python
df4.car_name=df4.car_name.apply(lambda x: 'other' if x in car_count_only_one else x)
len(df4.car_name.unique())
```

```python
df5=df4.drop(['km_driven'],axis='columns')
df5.head()
```

**APPLYING FLOAT FUNCTION for Kilometers_driven**

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df5[df5['kilometers_driven'].apply(is_float)].head()
```

**Making dummy variables of fuel type**

```python
dummies=pd.get_dummies(df5.fuel_type)
dummies
```

**adding the dummy variables with existing dataframe and creating a new dataframe**

```python
df6=pd.concat([df5,dummies],axis='columns')
df6.head()
```

**Applying dummy variable trap**

```python
df7=df6.drop(['Petrol'],axis='columns')
df7.head()
```

**using dummy variables for car names**

```python
new_dummies=pd.get_dummies(df7.car_name)
new_dummies
```

****creating a new dataframe by adding the dummy variables with existing dataframe****

```python
df8=pd.concat([df7,new_dummies.drop('other',axis='columns')],axis='columns')
df8.head()
```

```python
df9=df8.drop(['car_name'],axis='columns')
df9.head()
```

```python
df10=df9.drop(['fuel_type'],axis='columns')
df10.head()
```

**BUILDING THE MODEL**

```python
X=df10.drop(['price'],axis='columns')
X.head()
```

```python
y=df10.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

**Using Linear Regression measuring the accuracy of the model**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

**Testing the model with few properties**

```python
def predict_price(car_name,year,km,fuel):    
    car_index = np.where(X.columns==car_name)[0][0]
    fuel_index = np.where(X.columns==fuel)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = year
    x[1] = km
    if car_index >= 0:
        x[car_index] = 1 
    if fuel_index >= 0:
        x[fuel_index] = 1
  
    
    return model.predict([x])[0]
  
    
```

**Giving car name, year(model),kilometers driven and fuel type as inputs. based on these the model predicts the resale value of the car**

```python
predict_price('Volvo S90',2016,1000,'CNG')
```
